```python
m=heart.copy()
m=heart.drop('HeartDisease',axis=1)
```

```python
n=heart['HeartDisease']
```

```python
m_train,m_test,n_train,n_test=train_test_split(m,n,test_size=0.3,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
```

```python
a_score=[]
for k in range(7,18,2):
  knn=KNeighborsClassifier(n_neighbors=k)
  knn.fit(m_train,n_train)
  n_pred=knn.predict(m_test)
  accuracy=accuracy_score(n_test,n_pred)
  a_score.append(accuracy)

mse=[1 - m for m in a_score]
```

```python
max=a_score[0]
for j in a_score:
  if(j>max):
    max=j
print(max)
```

***RANDOM FOREST***

```python
heart
```

```python
sex=pd.get_dummies(heart['Sex'],drop_first=True)
st_slope=pd.get_dummies(heart['ST_Slope'],drop_first=True)

resting_ecg=pd.get_dummies(heart['RestingECG'],drop_first=True)
ex_angina=pd.get_dummies(heart['ExerciseAngina'],drop_first=True)
cpt=pd.get_dummies(heart['ChestPainType'],drop_first=True)
age=heart['Age']
hd=heart['HeartDisease']
```

```python
new_data=pd.concat([sex,age,st_slope,cpt,resting_ecg,ex_angina,hd],axis=1)
```

```python
new_data.head()
```

```python
a=new_data.HeartDisease
a.head()
```

```python
new_data.drop('HeartDisease',inplace=True,axis=1)
```

```python
new_data.head()
```

```python
b=new_data
```

```python
b_train,b_test,a_train,a_test=train_test_split(b,a,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(criterion='gini',n_estimators=20,random_state=0)
classifier.fit(b_train,a_train)
```

```python
a_pred=classifier.predict(b_test)
a_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report,accuracy_score
```

```python
cm=confusion_matrix(a_test,a_pred)
print(cm)
```

```python
accuracy_score(a_test,a_pred)
```

```python
print(classification_report(a_test,a_pred))
```
